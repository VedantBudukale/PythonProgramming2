```python
### Negative of the image
[rows,cols] = gray_img2.shape;
neg_img = np.zeros((rows,cols));
for r in range(0, rows-1):
  for c in range(0, cols-1):
    neg_img[r,c] = 255-gray_img2[r,c]; #Calculate negative image
#plt.imshow(neg_img, cmap=plt.cm.gray)
fig = plt.figure()
ax1 = fig.add_subplot(3,3,1)
ax1.imshow(img2)
ax1.set_title('Original image')
ax2 = fig.add_subplot(3,3,2)
ax2.imshow(gray_img2, cmap=plt.cm.gray)
ax2.set_title('Gray image')
ax3 = fig.add_subplot(3,3,3)
ax3.imshow(neg_img)
ax3.set_title('Negative image')
```
